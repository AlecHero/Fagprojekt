```python
# For each tsv file in the folder 'sleep_scores' with columns 'session', 'epoch_start_time' and '30-sec_epoch_sleep_stage', 
# visualize the hypnogram and save in folder hypnograms with a subplot for each session name
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
os.makedirs('hypnograms', exist_ok=True)
for subject in range(33):
    if subject+1 < 10:
        df = pd.read_csv('sleep_scores/sub-0'+str(subject+1)+'-sleep-stage.tsv', sep='\t')
    else:
        df = pd.read_csv('sleep_scores/sub-'+str(subject+1,)+'-sleep-stage.tsv', sep='\t')
    
    sessions = df['session'].unique()
    fig,axs = plt.subplots(len(sessions), 1,figsize=(5,3*len(sessions)),layout='constrained')
    for i,session in enumerate(sessions):
        session_df = df[df['session'] == session]
        # map W to W, 1 to N1, 2 to N2, 3 to N3
        session_df['sleep_stage'] = session_df['30-sec_epoch_sleep_stage'].map({'W':0, '1':1, '2':2, '3':3,'W (uncertain)':0, '1 (uncertain)':1, '2 (uncertain)':2, '3 (uncertain)':3})
        axs[i].plot(session_df['epoch_start_time_sec']/60, session_df['sleep_stage'])
        axs[i].set_title('Hypnogram for '+session)
        axs[i].set_xlabel('Time [min]')
        axs[i].set_ylabel('Sleep stage')
        axs[i].set_yticks([0,1,2,3])
        axs[i].set_yticklabels(['W', 'N1', 'N2', 'N3'])
        #reverse axis
        axs[i].invert_yaxis()
    plt.savefig('hypnograms/sub-'+str(subject+1)+'_hypnogram.png')
    plt.close()
```
